```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from data_collection import generate_parameters, paper_experiments
from means_and_covars import means_opts, covars_opts
```

### Testing results

```python
MSE_Random = list()
MSE_Grid = list()
MSE_LH = list()
MSE_Sob = list()
MSE_Hal = list()
KLD_Random = list()
KLD_Grid = list()
KLD_LH = list()
KLD_Sob = list()
KLD_Hal = list()
```

```python
vals_cols = [MSE_Random, MSE_Grid, MSE_LH, MSE_Sob, MSE_Hal, KLD_Random, KLD_Grid, KLD_LH, KLD_Sob, KLD_Hal]
```

```python
n_epochs = 15
n_dimensions = [2]
n_hidden = 2
sample_sizes = [300]
grid_sizes = [6, 8]
for n_dimension in n_dimensions:
    print(f'Doing dimension {n_dimension}')
    _, _, startprob, transmat = generate_parameters(n_dimension, n_hidden)
    for sample_size in sample_sizes:
        for grid_size in grid_sizes:
            losses_all, scores_all, acc_all, discr_all, mod_opt = paper_experiments(sample_size,
                                                                                    grid_size,
                                                                                    n_dimension,
                                                                                    n_hidden,
                                                                                    means_opts[f'means{n_hidden}_{n_dimension}'],
                                                                                    covars_opts[f'covars{n_hidden}_{n_dimension}'],
                                                                                    startprob,
                                                                                    transmat, n_epochs)
            MSE_Random.append([losses_all[0], scores_all[0], acc_all[0], discr_all[0]])
            MSE_Grid.append([losses_all[1], scores_all[1], acc_all[1], discr_all[1]])
            MSE_LH.append([losses_all[2], scores_all[2], acc_all[2], discr_all[2]])
            MSE_Sob.append([losses_all[3], scores_all[3], acc_all[3], discr_all[3]])
            MSE_Hal.append([losses_all[4], scores_all[4], acc_all[4], discr_all[4]])
            KLD_Random.append([losses_all[5], scores_all[5], acc_all[5], discr_all[5]])
            KLD_Grid.append([losses_all[6], scores_all[6], acc_all[6], discr_all[6]])
            KLD_LH.append([losses_all[7], scores_all[7], acc_all[7], discr_all[7]])
            KLD_Sob.append([losses_all[8], scores_all[8], acc_all[8], discr_all[8]])
            KLD_Hal.append([losses_all[9], scores_all[9], acc_all[9], discr_all[9]])
```
